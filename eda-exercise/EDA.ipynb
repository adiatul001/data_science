**Treating outliers in `Prevailing Wage` column**

```python
df['PREVAILING_WAGE'].describe()
```

```python
sns.boxplot(data=df, x='PREVAILING_WAGE', orient='h')
plt.show()
```

```python
median = df['PREVAILING_WAGE'].median()
Q1 = df['PREVAILING_WAGE'].quantile(0.25)
Q3 = df['PREVAILING_WAGE'].quantile(0.75)
IQR = Q3-Q1

lower = Q1 - 1.5*IQR
upper = Q3 +1.5*IQR

df['PREVAILING_WAGE'] = np.where((df['PREVAILING_WAGE'] < lower) | (df['PREVAILING_WAGE'] > upper), median, df['PREVAILING_WAGE'])
```

<h4 style="font-family: monospace">8. Rate of denial of visa petition by the states</h4>

```python
df_temp = df[['STATE_NAME', 'CASE_STATUS']]
```

```python
df_temp.head()
```

```python
df_denied = df_temp[df_temp['CASE_STATUS'] == 'DENIED'].groupby('STATE_NAME').size().reset_index(name='count')
```

```python
df_denied.head()
```

```python
df_denied['percentage'] = (df_denied['count'] / df_denied['count'].sum())*100

for index, percentage in df_denied['percentage'].items():
    if percentage <= 1.5:
        df_denied.loc[index, 'STATE_NAME'] = 'others'
```

```python
df_denied.head()
```

```python
df_others = df_denied[df_denied['STATE_NAME'] == 'others']
others_count = df_others['count'].sum()
others_percentage = df_others['percentage'].sum()
print(f'{others_count} and {others_percentage}')

df_denied = df_denied[df_denied['STATE_NAME'] != 'others']

df_others = pd.DataFrame({
    'STATE_NAME': ['others'],
    'count': [others_count.item()],
    'percentage': [others_percentage.item()]
})
new_df_denied = pd.concat([df_denied, df_others])
new_df_denied
```

```python
explode = [0.1 if percentage < 5else 0 for percentage in df_denied['percentage']]


figure = plt.figure(figsize=(16, 16))
sns.set_palette('rainbow')
sns.set_style('whitegrid')

plt.pie(df_denied['count'], labels=df_denied['STATE_NAME'], autopct='%1.1f%%', explode=explode, pctdistance=0.8,
       labeldistance=1.1)
plt.title('Denied VISA petitions by states', pad=30)
plt.axis('equal')
plt.show()
```
